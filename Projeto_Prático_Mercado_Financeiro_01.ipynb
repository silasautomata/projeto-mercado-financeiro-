```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_excel('/content/Vase_004 - Magalu - Sem Resolução (1).xlsx')

base_dados.head()
```

```python
# series temporais
dados = base_dados.set_index('Data') # retirando o index e ordenado o index pela data
# ordenado de inicio ao fim de ano de 2021
dados.head()
```

```python
dados.tail()
```

```python
# grafico plot do das açoes de 2021 a 2022

plt.figure(figsize=(16,5)) # tamando da figura do grid
plt.title('analise das açoes da magalu - Fechamento', fontsize=15, loc='left') # configuraçao do titulo
plt.plot(dados.index,dados['Fechamento']) # gera o tipod e grafico com a coluna de fechamento do inicioa de 20021 a final de 2021

plt.xlabel('periodo da coraçao') # titulo do eixo X
plt.ylabel('valor de açao ($)'); # titulo do eixo Y
```

```python
dados.tail()
```

```python
media_movel = dados['Fechamento'].rolling(5).mean()# gera a media movel de 5 dias
media_tendencia = dados['Fechamento'].rolling(30).mean()# gera a media de tendencia de 30 dias
```

```python
# gerando grafico e da media movel e da media de tendencia
plt.figure(figsize=(16,5))
plt.title('analise das açoes da magalu - Fechamento', fontsize=15, loc='left')

plt.plot(dados.index,dados['Fechamento']) #
plt.plot(dados.index,media_movel) #
#plt.plot(dados.index,media_tendencia)#
```

```python
plt.plot(dados.index,dados['Fechamento'])
plt.plot(dados.index,media_movel)
plt.plot(dados.index,media_tendencia)
```

```python
plt.xlabel('periodo da coraçao')
plt.ylabel('valor da açao (R$)');
```

```python
media_movel = dados['Fechamento'].rolling(5).mean()
media_tendencia = dados['Fechamento'].rolling(30).mean()

#plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')

plt.plot( dados.index, dados['Fechamento'] ) # linha azul
plt.plot( dados.index, media_movel ) # media movel = linha laranja
plt.plot( dados.index, media_tendencia ) # media de tendencia = linha verde

plt.xlabel('Período da Coração')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot(data=dados, x='Fechamento')
```

```python
# gera uma coluna 'mes
base_dados['Mes'] = base_dados['Data'].dt.month

plt.figure(figsize=(16,5))
sns.boxplot(data=base_dados, x='Mes', y='Fechamento');
```

```python
base_dados.groupby(['Mes']).describe()['Fechamento']
```

```python
grafico = go.Figure(
    data=[
        go.Candlestick(
            x=dados.index,
            open = dados['Abertura'],
            high = dados['Maior'],
            low = dados['Menor'],
            close = dados['Fechamento'],
        )
    ]
)

grafico.update_layout(xaxis_rangeslider_visible=False)
grafico.show()
```
